```python
class AzureContainer:
    def __init__(self, base_path: str):
        """
        Inicializa o container apenas com o path WASBS.
        A autenticação já foi configurada no cluster
        """
        self.base_path = base_path.rstrip("/") + "/"

    def list_files(self, path: str = "", mostrar: bool = True):
        """Lista arquivos e opcionalmente exibe no Databricks."""
        try:
            files = dbutils.fs.ls(self.base_path + path)
            if mostrar:
                display(files)
            return files
        except Exception as e:
            print(f"Erro ao listar arquivos em {self.base_path}: {e}")
            return []

    def get_files_list(self, path: str = ""):
        """Retorna a lista de arquivos para uso programático."""
        try:
            return dbutils.fs.ls(self.base_path + path)
        except Exception as e:
            print(f"Erro ao acessar {self.base_path}: {e}")
            return []
```
