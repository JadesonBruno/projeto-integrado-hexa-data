```python
class AzureContainer:
    def __init__(self, storage_account, container_name, sas_token):
        self.storage_account = storage_account
        self.container_name = container_name
        self.sas_token = sas_token
        self.base_path = f"wasbs://{self.container_name}@{self.storage_account}.blob.core.windows.net/"
        
        # Configura o Spark automaticamente ao instanciar o objeto
        self._configure_spark()

    def _configure_spark(self):
        """Configura a autenticação para este container específico."""
        conf_key = f"fs.azure.sas.{self.container_name}.{self.storage_account}.blob.core.windows.net"
        spark.conf.set(conf_key, self.sas_token)

    def list_files(self, path="", mostrar=True):
        """Retorna a lista de arquivos e opcionalmente exibe a tabela."""
        try:
            files = dbutils.fs.ls(self.base_path + path)
            if mostrar:
                display(files)
            return files
        
        except Exception as e:
            print(f"Erro ao listar arquivos em {self.container_name}: {e}")
            return []
        
    def get_files_list(self, path=""):
        """Retorna a lista de objetos de arquivos do dbutils."""
        try:
            return dbutils.fs.ls(self.base_path + path)
        
        except Exception as e:
            print(f"Erro ao acessar {self.container_name}: {e}")
            return []
```
