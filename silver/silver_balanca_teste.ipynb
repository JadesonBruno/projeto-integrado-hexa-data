```python
%sql
SHOW TABLES IN silver_balancacomercial;
```

```python
 %sql
 SELECT 
     table_name,
     last_ingestion_timestamp,
     last_update,
     records_processed,
     execution_id
 FROM metadata.silver_control_balanca
 ORDER BY last_update DESC;
```

```python
 %sql
 SELECT 
     ano,
     mes,
     COUNT(*) as total_registros,
     SUM(vl_fob) as valor_total_fob,
     SUM(kg_liquido) as peso_total_kg
 FROM silver_balancacomercial.silver_exp
 GROUP BY ano, mes
 ORDER BY ano DESC, mes DESC
 LIMIT 12;
```

```python
 %sql
 SELECT 
     p.no_pais,
     SUM(e.vl_fob) as valor_total
 FROM silver_balancacomercial.silver_exp e
 JOIN silver_balancacomercial.silver_pais p ON e.co_pais = p.co_pais
 WHERE e.ano = 2021
 GROUP BY p.no_pais
 ORDER BY valor_total DESC
 LIMIT 10;
```
